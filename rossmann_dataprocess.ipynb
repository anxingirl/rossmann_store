## 3.进行特征工程

```python
train_feature= pd.read_csv('train_store.csv',parse_dates=[12],low_memory=False)
test_feature  = pd.read_csv('test_store.csv',parse_dates=[11],low_memory=False)
```

```python
train_feature.head()
```

```python
test_feature.columns
```

```python
def feature_project(data):
    print("StoreType, Assortment, StateHoliday都是OBJECT类型，将这三个字段转变为数字类型。")
    #将存在其他字符表示分类的特征转化为数字
    #mappings = {'0':0, 'a':1, 'b':2, 'c':3, 'd':4}
    #data.StoreType.replace(mappings, inplace=True)
    #data.Assortment.replace(mappings, inplace=True)
    #data.StateHoliday.replace(mappings, inplace=True)
    #强制将StateHoliday数据转化为INT类型
    #data['StateHoliday'] = data['StateHoliday'].astype(int)
    #强制将Assortment数据转化为INT类型
    #data['Assortment'] = data['Assortment'].astype(int)
    #强制将StoreType数据转化为INT类型
    #data['StoreType'] = data['StoreType'].astype(int)
    
    print("年份修改")
    data['Year'] = data.Date.dt.year
    data['Month'] = data.Date.dt.month
    data['Day'] = data.Date.dt.day
    data['DayOfWeek'] = data.Date.dt.dayofweek
    data['WeekOfYear'] = data.Date.dt.weekofyear
    
    print("取出竞争对手已经开业的时间")
    #取出竞争对手已经开业的时间
    #推广活动开展了多长时间，如果PROMO2为0，则直接设置为0
    data['CompetitionOpen'] = 12 * (data.Year - data.CompetitionOpenSinceYear) + (data.Month - data.CompetitionOpenSinceMonth)
    data['CompetitionOpen'] = data.CompetitionOpen.apply(lambda x: x if x > 0 else 0)  
    data['Promo2Open'] = data.Promo2 * (12 * (data.Year - data.Promo2SinceYear) + (data.WeekOfYear - data.Promo2SinceWeek) / 4.0)
    data['Promo2Open'] = data.Promo2Open.apply(lambda x: x if x > 0 else 0)
    
    print("设置月份")
    month2str = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sept', 10:'Oct', 11:'Nov', 12:'Dec'}
    data['monthStr'] = data.Month.map(month2str)

    data['IsPromoMonth'] = 0 
    data.loc[data.PromoInterval == 0, 'PromoInterval'] = ''
    for interval in data.PromoInterval.unique():

        if interval != '':

            for month in interval.split(','):

                data.loc[(data.monthStr == month) & (data.PromoInterval == interval), 'IsPromoMonth'] = 1
    print("删除没有开业的时间")
    ### 删除掉没有开业的时间
    #data.drop(data[data.Open == 0].index, inplace = True)
    #print("删除掉销量为0的时间")
    #data.drop(data[data.Sales == 0].index, inplace = True)
    print("删除promointerval ,monthstr")
    ### 删除掉没有价值的字符串序列
    data = data.drop(['PromoInterval','monthStr'],axis = 1, inplace = True) 
    
```
